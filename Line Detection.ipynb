```python
img = cv2.imread('img/1.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
edges = cv2.Canny(np.uint8(gray),50,150,apertureSize = 3)
cv2.imwrite('edges-50-150.jpg',edges)
lines = cv2.HoughLines(edges,1,np.pi/180, 200) 
plt.imshow(edges)
```

```python
for r,theta in lines[0]: 
    a = np.cos(theta) 
    b = np.sin(theta) 
    x0 = a*r 
    y0 = b*r 
    x1 = int(x0 + 10000*(-b)) 
    y1 = int(y0 + 10000*(a)) 
    x2 = int(x0 - 10000*(-b)) 
    y2 = int(y0 - 10000*(a)) 
    cv2.line(img,(x1,y1), (x2,y2), (255,0,0),20) 
plt.imshow(img)
```

```python
img2 = img.copy()
img2[:y1 + 20,:] = 0
plt.imshow(img2)
```
